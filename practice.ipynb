Check whether a given number is a prime number or not

```python
def primecheck(n):
    if n<=1:
        return False
    
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
        
    return True

num = int(input('Enter an integer: '))
val = primecheck(num)

if val:
    print(f'The number {num} is a prime number')
else:
    print(f"The number {num} is not a prime number")
```

You are given a list of academic records. Each record is a Tuple in the format:
(Department, Student_Name, Subject, Score).
Because the data comes from different faculty databases, there are duplicate entries that
need to be cleared. Your goal is to process this data to provide a summary of scores grouped
by department.
Requirements:
I. Deduplicate: Remove any identical records using a Set.
II. Group: Use a defaultdict to group all scores (integers) under their respective
Department keys.
III. Aggregate: Create a final Dictionary where the key is the Department and the value
is a Tuple containing:
a. The total number of unique students in that department.
b. The average score of the department.
IV. Order: Ensure the final output is printed in alphabetical order of the departments.

```python
from collections import defaultdict

record = [
    ("CS", "Alice", "Python", 90),
    ("Math", "Bob", "Calculus", 80),
    ("CS", "Alice", "Python", 90),
    ("CS", "Charlie", "C++", 70),
    ("Math", "Bob", "Algebra", 70),
    ("Physics", "David", "Optics", 85),
    ("CS", "Alice", "Algorithms", 95)
]

#1
uniRecords = set(record)

#2
deptScore = defaultdict(list)
deptStd = defaultdict(set)

for dept, name, sub, score in uniRecords:
    deptScore[dept].append(score)
    deptStd[dept].add(name)

#3
finaldict = {}

for dept in deptScore:
    avgScore = sum(deptScore[dept])/len(deptScore[dept])
    totalStd = len(deptStd[dept])
    finaldict[dept] = (totalStd, avgScore)

#4
print(f'Unique records of students: {uniRecords}\n')
print('Department  Unique Student  Average Score\n')

for dept in sorted(finaldict.keys()):
    std, avg = finaldict[dept]
    print(f'{dept}\t\t{std}\t\t{avg}\n')
```

```python
import sys

class Account:
    """Represents a customer's individual bank account."""
    def __init__(self, account_number, pin, account_holder, initial_balance=0.0):
        self.account_number = account_number
        self.__pin = pin  # Encapsulation: Private variable to secure the PIN
        self.account_holder = account_holder
        self.__balance = initial_balance  # Encapsulation: Protected balance access

    def verify_pin(self, entered_pin):
        """Verifies if the entered PIN matches the account PIN securely."""
        return self.__pin == entered_pin

    def get_balance(self):
        """Getter method to securely check the balance."""
        return self.__balance

    def deposit(self, amount):
        """Adds funds to the account if the amount is valid."""
        if amount > 0:
            self.__balance += amount
            return True, f"Successfully deposited ${amount:.2f}"
        return False, "Invalid deposit amount."

    def withdraw(self, amount):
        """Deducts funds if the balance is sufficient."""
        if amount <= 0:
            return False, "Invalid withdrawal amount."
        if amount > self.__balance:
            return False, "Insufficient funds available."
        
        self.__balance -= amount
        return True, f"Successfully withdrew ${amount:.2f}"


class BankDatabase:
    """Simulates the bank's central server holding customer accounts."""
    def __init__(self):
        # Initializing the database with dummy customer records
        self._accounts = {
            "1001": Account("1001", "1234", "Alice Smith", 5000.00),
            "1002": Account("1002", "5678", "Bob Jones", 1200.50),
            "1003": Account("1003", "9999", "Charlie Brown", 50.00)
        }

    def get_account(self, account_number):
        """Retrieves an account object if it exists."""
        return self._accounts.get(account_number, None)


class ATM:
    """Handles the physical hardware interaction and session state machine."""
    def __init__(self, bank_database):
        self.db = bank_database
        self.current_account = None  # Tracks who is currently logged in

    def start(self):
        """The main execution loop for the ATM hardware interface."""
        print("=== WELCOME TO APEX COOPERATIVE BANK ATM ===")
        acc_num = input("Please enter your Account Number: ").strip()
        account = self.db.get_account(acc_num)

        if not account:
            print("Error: Account record not found. Card returned.")
            return

        # PIN Validation Layer (Max 3 attempts simulation)
        attempts = 0
        while attempts < 3:
            pin = input("Enter your 4-digit PIN: ").strip()
            if account.verify_pin(pin):
                self.current_account = account
                print(f"\nWelcome back, {account.account_holder}!")
                self._show_menu()
                return
            else:
                attempts += 1
                print(f"Incorrect PIN. Attempts remaining: {3 - attempts}")

        print("\nCard confiscated due to excessive wrong attempts. Please contact your branch.")

    def _show_menu(self):
        """Internal system menu for authenticated users."""
        while True:
            print("\nATM options:")
            print("1. Check Balance")
            print("2. Deposit Funds")
            print("3. Withdraw Cash")
            print("4. Exit")
            
            choice = input("Select an option (1-4): ").strip()

            if choice == "1":
                print(f"Your Current Balance is: ${self.current_account.get_balance():.2f}")
            
            elif choice == "2":
                try:
                    amount = float(input("Enter deposit amount: $"))
                    success, message = self.current_account.deposit(amount)
                    print(message)
                except ValueError:
                    print("Error: Please enter a valid numerical value.")

            elif choice == "3":
                try:
                    amount = float(input("Enter withdrawal amount (Multiples of $20): $"))
                    success, message = self.current_account.withdraw(amount)
                    print(message)
                except ValueError:
                    print("Error: Please enter a valid numerical value.")

            elif choice == "4":
                print("\nThank you for banking with us. Please take your card! Goodbye.")
                self.current_account = None
                break
            else:
                print("Invalid configuration option selected. Please try again.")


# --- Execution Block ---
if __name__ == "__main__":
    # Create the central database
    central_db = BankDatabase()
    
    # Initialize the ATM machine with the central database access
    atm_machine = ATM(central_db)
    
    # Fire up the machine interface
    atm_machine.start()
```

```python
class Shop:
    def __init__(self):
        self.__cart = []
    
    def addProd(self, items, price, quantity):
        self.__cart.append((items, price, quantity))
    
    def __calc_tot(self):
        total = 0
        for items, price, quantity in self.__cart:
            total += price*quantity
        return total
    
    def show_bill(self):
        for item, price, quantity in self.__cart:
            print(f'Your bought {quantity} number of {item} of price {price}')
        
        print(f'Total Bill: {self.__calc_tot()}')

c = Shop()
while True:
    item = input('Enter item you want to buy: ')
    price = float(input('Enter the price: '))
    qty = float(input('Enter quantity: '))
    
    c.addProd(item,price,qty)
    choice = input('Are you done?').lower().strip()
    if choice=='no':
        break

c.show_bill()
```

An online shopping platform allows customers to pay using different payment methods such as Credit Card and Bitcoin. Although all payment methods must
perform the common task of processing a payment.  the actual implementation differs: A Credit Card payment requires validating the card number and
charging the amount. A Bitcoin payment requires verifying the wallet address and transferring cryptocurrency. The checkout system should not need to know the internal details of each payment method. It should simply call a common method to process the payment. Implement this system using Data Abstraction in Python by creating an abstract
class PaymentMethod with an abstract method process_payment(). Then create concrete subclasses CreditCard and Bitcoin that provide their own
implementations.

```python
from abc import ABC, abstractmethod

class PaymentMethod(ABC):
    @abstractmethod
    def processPayment(self, amount):
        pass

class Database:
    def __init__(self):
        self._account = ['1001-102-101', '1001-102-102', '1001-102-103', '1001-102-104']
        self._bitid = ['A1B2C3D4F1', 'A1B2C3D4F2', 'A1B2C3D4F3', 'A1B2C3D4F4', 'A1B2C3D4F5']

class CreditCard(PaymentMethod):
    def __init__(self, acc, db):
        self.acc = acc
        self.db = db

    def processPayment(self, amount):
        print("Credit card payment...")
        if self.acc in self.db._account:
            print('Card validated')
            print(f'Processing payment of {amount} TK')
        else:
            print('Enter a valid account number')

class Bitcoin(PaymentMethod):
    def __init__(self, acc, db):
        self.acc = acc
        self.db = db

    def processPayment(self, amount):
        print("\nBitcoin payment...")
        if self.acc in self.db._bitid:
            print('Bitcoin ID validated')
            print(f'Processing payment of {amount} Bitcoin')
        else:
            print('Enter a valid Bitcoin ID')

db=Database()
c = CreditCard('1001-102-102', db)
b = Bitcoin('A1B2C3D4F3', db)

c.processPayment(10000)
b.processPayment(20)

```

Write a python function called extract_middle() that takes a list of integers and returns a new list containing the middle half of the elements. The function should return the third quartile integers of the list.

```python
def extract_middle(num):
    n = len(num)
    
    start = int(n*0.5)
    end = int(n*0.75)
    return num[start:end]

num = [10,10,24,13,28,28,20,12,15,24,18,19,30,21,25]
num = sorted(num)
print('The original number: ',num)
newlist = extract_middle(num)
print('The middle part of the number: ',newlist)
```

Develop a python program to classify students into the following grade categories: "A" (80 and above), "B" (70-79), "C" (60-69) and "Fail" (below 60). Store and present the classified result in a structured dictionary grouped by grade

```python
students = [
    {'id':1, 'name':'A', 'marks':78},
    {'id':2, 'name':'B', 'marks':91},
    {'id':3, 'name':'C', 'marks':67},
    {'id':4, 'name':'D', 'marks':50},
    {'id':5, 'name':'E', 'marks':75}
]

print(students)

group = {'A':[], 'B':[], 'C':[], 'F':[]}

for std in students:
    m = std['marks']
    if m >= 80:
        grade='A'
    elif m >= 70 and m<=79:
        grade='B'
    elif m >= 60 and m<=69:
        grade='C'
    else:
        grade='F'
    print(f'Student name: {std['name']} Marks: {std['marks']} Grade: {grade}')
    
    group[grade].append({
        'id':std['id'], 'name':std['name'], 'marks':m
    })

print("Result grouped by grade: ",group)
```

You are given a list of integers representing daily temperatures. A peak is defined as a value that is strictly greater than its immediate neighbours. Your task is to I)Identify all the peaks in the list, II) Print the index and value of each peak, III) Print the total number of peaks

```python
temps = [30,32,29,35,34,36,33,31,40,39]
l = len(temps)
index = []
for i in range(l-1):
    if i==0 and temps[i]>temps[i+1]:
        index.append(i)
    elif i==l-1 and temps[i]>temps[i-1]:
        index.append(i)
    elif temps[i]>temps[i+1] and temps[i]>temps[i-1]:
        index.append(i)

print('The peaks are at index: ',index)
for i in index:
    print(f'Peak found at index {i} with value {temps[i]}')

print('Total number of peaks: ',len(index))
```

Design a library management system where define a base class LibraryItem and derive classes Book and CD from it. Include appropriate attributes and demonstrate method overriding where applicable. Create instances of the classes and implement a method to display all issued items in the system.

```python
class LibraryItem:
    def __init__(self, name, year, status):
        self.name = name
        self.year = year
        self.__status = status
    
    def borrow(self):
        print('The item is issued')

    def ret(self):
        print('The item is returned')
    
    def display(self):
        print('Name: ',self.name)
        print('Year: ',self.year)
    
    def get_status(self):
        return self.__status
    
    def set_status(self,status):
        self.__status = status

class Book(LibraryItem):
    def __init__(self,name,year,status,author):
        super().__init__(name,year,status)
        self.author = author
    
    def borrow(self):
        print('The book is issued')
        super().set_status('Issued')

    def ret(self):
        print('The book is returned')
        super().set_status('Available')

class CD(LibraryItem):
    def __init__(self,name,year,status,publisher):
        super().__init__(name,year,status)
        self.publisher = publisher
    
    def borrow(self):
        print('The CD is issued')
        super().set_status('Issued')

    def ret(self):
        print('The CD is returned')
        super().set_status('Available')

def displayIssue(obj):
    if obj.get_status() == 'Issued':
        obj.display()

b1 = Book('ABC',2008,'Issued','MNO')
b2 = Book('XYZ',1998,'Available','RO')
c1 = CD('JHK',2010,'Issued','KOL')
c2 = CD('MAF',2020,'Issued','OIP')

b1.ret()
b2.borrow()

for i in [b1,b2,c1,c2]:
    displayIssue(i)
```

Write a python program that demonstrates file handling and exception handling by asking the user to input a filename, attempting to open and read the file, displaying its contents if the file exists and handling errors such as file not found or invalid file access using appropriate exception handling techniques

```python
filename = input('Enter a filename: ')
try:
    with open(filename, 'r') as file:
        data = file.read()
    print("Content: ",data)
except FileNotFoundError as f:
    print('Error: ',f)
except PermissionError as p:
    print('Error: ',p)
except Exception as e:
    print('Unexpected error: ',e)
else:
    print('Execution complete')
```
